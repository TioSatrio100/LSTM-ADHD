```python
# Pre-processing
def butterworth_filter(signal, lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    filtered_signal = filtfilt(b, a, signal)
    return filtered_signal

# Filter sinyal dengan bandpass filter pada df_subset
fs = 200
lowcut = 1
highcut = 50

numeric_columns = df_subset.select_dtypes(include=[np.number]).columns
data_filtered_subset = df_subset[numeric_columns].copy()
for column in numeric_columns:
    data_filtered_subset[column] = butterworth_filter(data_filtered_subset[column].values, lowcut, highcut, fs)
```
